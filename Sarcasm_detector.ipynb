```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/Sarcasm Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0,3468):
  review = re.sub('^[a-zA-Z]',' ',str(dataset['tweet'][i]))
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in all_stopwords]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
x = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
from sklearn import svm

classifier = svm.SVC(kernel = 'rbf')
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)

print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm = confusion_matrix(y_test,y_pred)
print(cm)

accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB

classifier2 = GaussianNB()

classifier2.fit(x_train,y_train)
```

```python
y_pred2 = classifier2.predict(x_test)
print(y_pred2)
```

```python
cm = confusion_matrix(y_test,y_pred2)
print(cm)

accuracy_score(y_test,y_pred2)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = [LogisticRegression(),svm.SVC(kernel='rbf'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def cross_val_compare():
  for model in models:
    cv_score = cross_val_score(model,x,y,cv=5)
    mean_accuracy = sum(cv_score)/len(cv_score)
    mean_accuracy = mean_accuracy*100
    mean_accuracy = round(mean_accuracy,2)

    print("Cross Validation Accuracy for the ",model,"= ",cv_score)
    print("Accuracy % for the ",model,"= ",mean_accuracy)
```

```python
cross_val_compare()
```

```python
cv_score_nb = cross_val_score(GaussianNB(),x,y,cv=5)
print(cv_score_nb)
mean_accuracy_nb = sum(cv_score_nb)/len(cv_score_nb)
mean_accuracy_nb = mean_accuracy_nb*100
mean_accuracy_nb = round(mean_accuracy_nb,2)
print(mean_accuracy_nb)
```

```python
def calculate_model_weight(error):
  return 0.5*np.log((1-error)/(error))
```

```python
alpha1 = calculate_model_weight(0.161)
alpha1
```

```python
def update_row_weights(row,alpha = 0.825):
  if row['sarcastic'] == row['y_pred']:
    return row['weights']*np.exp(-alpha)
  else:
    return row['weights']*np.exp(alpha)
```

```python
dataset['updated_weights'] = dataset.apply(update_row_weights,axis = 1)
dataset.head()
```

```python
dataset['normalised_weights'] = dataset['updated_weights']/dataset['updated_weights'].sum()
```

```python
dataset['normalised_weights'].sum()
```

```python
dataset['cumsum_upper'] = np.cumsum(dataset['normalised_weights'])
```

```python
dataset['cumsum_lower'] = dataset['cumsum_upper'] - dataset['normalised_weights']
```

```python
def create_new_dataset(dataset):
  indices = []

  for i in range(dataset.shape[0]):
    a = np.random.random()
    for index,row in dataset.iterrows():
      if a >= row['cumsum_lower'] and a <= row['cumsum_upper']:
        indices.append(index)
  return indices
```

```python
index_values = create_new_dataset(dataset)
```

```python
second_df = dataset.iloc[index_values,[0,1,2,9]]
```
